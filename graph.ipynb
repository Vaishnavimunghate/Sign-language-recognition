```python
# GRAPH 1
import matplotlib.pyplot as plt
import numpy as np

# Define x and y values
x1 = np.array([96.31, 96.30, 96.37, 96.26, 96.06, 96.00])
x2 = np.array([74.34, 74.81, 74.65, 73.78, 73.54, 72.91])
x3 = np.array([88.74, 89.01, 89.27, 88.61, 89.27, 88.74])
y = np.array([200, 500, 700, 1000, 1200, 1600])

# Create plot
plt.plot(y, x1, 'ro-', label='ASL Alphabet')
plt.plot(y, x2, 'bs-', label='Sign Numbers')
plt.plot(y, x3, 'g^-', label='Massey')
plt.ylabel('Accuracy')
plt.xlabel('Principal Component')
plt.legend()
plt.show()
```
